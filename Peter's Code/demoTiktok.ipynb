```python
import requests
import urllib.request
from urllib.parse import urlparse, urljoin
import time
import numpy as np


#make sure url valid

def is_valid(url):
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)
    
    
    
```

```python
def scrape_data(url):
#url = messages[5]

#     parsed = urlparse(url)
#     # remove URL GET parameters, URL fragments, etc.
#     url = parsed.scheme + "://" + parsed.netloc + parsed.path

#     if not is_valid(url):
#         data = " "
#     else:
        response = requests.get(url)
        response_page = response.content
        soup = BeautifulSoup(response.text, "html.parser")
        text = soup.find_all(text=True)
        #output = np.empty([])
        #output = pd.Series(data=data)
        
        
#        """
        data = []


        blacklist = [
            '[document]',
            'noscript',
            'header',
            'html',
            'meta',
            'head', 
            'input',
            'script',
            'body',
            'menu',
            'style',
            'tt-feed'
            #'jsx-195647250 header-container white middle',
            
            # there may be more elements you don't want, such as "style", etc.
        ]
        
        whitelist=[
            'jsx-1473136130 share-layout-header share-header',
            'number'
            
        ]

#        for t in text:
#            if (t.parent.name not in blacklist) or (t.parent.name in whitelist):
#                data += '{}'.format(t)

        for t in text:
            if (t.parent.name in blacklist):
                data += '{}'.format(t)



#        """
        


        #print("NOT APPENDED LIST")
        #print(output)
        data = "".join(data)
        
        print("APPENDED LIST\n")
        #print(data)

        return data

    
```

```python
#iterate through url addresses in each row, call scrape data with url, place text
#in 'text column of dataframe'

for urls in df["url"]:
   # print(scrape_data(urls))#test if scrape data working with correct values (urls)
    df["text"][urls] = scrape_data(urls)



  
```
